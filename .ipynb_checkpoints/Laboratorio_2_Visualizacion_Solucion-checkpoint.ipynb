# Laboratorio 2: Visualización

El análisis exploratorio de los datos (EDA) es uno de los primeros pasos en un proyecto de minería de datos luego de la comprensión del problema/negocio y la selección de los datos. Es de suma importancia ya que permite corroborar hipótesis, detectar outliers, estructurar el modelamiento de los datos, resolver problemas de los datos e iterar en el entendimiento del problema.

Algunos de los principales desafíos son como visualizar datos no estructurados (como texto), visualizar estructuras de datos complejas como redes y visualizar conjunto de datos de alta dimensionalidad.

## 1. Teórico, 2 ptos (0.5ptos c/u)

1.1 Por qué la visualización de datos es tan relevante en el análisis exploratorio de datos.

La visualización de datos juega un rol importante en el EDA (Exploratory Data Analysis), pues sirven para mostrar como se distribuyen los datos (dispersión/blox-plot, histogramas, gráficos de barra), medir correlación entre variables, detección de outliers, selección de característiccas. Además son capaces de mostrar una perspectiva que las estadísticas descriptivas no son capaces de mostrar (https://www.autodeskresearch.com/publications/samestats). 

1.2 En qué contexto debiese utilizar un gráfico de barras, cúando debería utilizar un gráfico de barras porcentual por sobre uno basado en frecuencia (de un ejemplo). 

El dominio de la variable es categórica, pues cuenta la ocurrencia de una categoría o valor que toma la variable en los datos. El gráfico porcentual se debiese utilizar cuando cruzamos dos variables categóricas, por ejemplo genero y fuga, si la variable genero esta desbalanceada, por ejemplo son más mujeres que hombres (1000 versus 500) y del total de 1000 fugas 600 corresponden a mujeres y 400 a hombres, entonces, la variable genero podría dar a entender que las mujeres se fugan más solo por el hecho que son más en la base de datos, en cambio, el histograma de barras porcentual normaliza la barra fuga|mujer (600) y fuga|hombre (400) por el tamaño del grupo mujer (1000) y hombre (500) respectivamente, por lo que un gráfico de barras porcentual entregaría la probabilidad de fuga dado que es mujer (600/1000=0.6) y la probabilidad de fuga dado que es hombre (400/500=0.8).

1.3 En qué contexto debiese utilizar un histograma, cúando debería utilizar un histograma de densidad por sobre uno basado en frecuencia (de un ejemplo). 

El dominio de la variable debe ser continua, el histograma discretiza una variable continua en intervlos (bins), luego realiza un conteo por intervalo de cuántas observaciones cayeron ahí. El gráfico percentual se debiese utilizar cuando cruzamos una variable continua con una categórica, por ejemplo la edad y fuga, si en la base de datos se tiene que de un total de 1100 personas tan solo 100 se fugaron, al comparar los histogramas condicionales (edad|fuga versus edad|no fuga) lo más probables es que se tenga que cada barra del histograma de la edad del grupo no fuga es más alto que el del grupo fuga, por tanto, es necesario normalizar por el tamaño de cada grupo obteniendo así histograma de densidades donde el área que cubren es el mismo e igual a 1, así podremos observar diferencias en la forma de la distribución y no en el tamaño.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
```

## Funciones

```python
def percentage_df(df, cat_col, label_col): 
    df_count = df.groupby([cat_col, label_col]).size().unstack(fill_value=0).stack().reset_index(name='count')
    df_count.sort_values([cat_col, label_col], inplace=True)

    cat_count = list(df_count.groupby(cat_col).sum()['count'])
    N = len(df_count[label_col].unique())
    df_count.loc[:, '%'] = 100*round(df_count['count']/np.array([[elem]*N for elem in cat_count]).reshape(len(df_count)), 2)
    return df_count
```

```python
def stacked_barplot(df_percentage, cat_col, label_col, label_order):
    sns.set()

    plt.figure(figsize=(8, 4))

    #stacked frequency barplot
    x1, y1 = list(df_percentage[df_percentage[label_col]==label_order[0]][cat_col]), list(df_percentage[df_percentage[label_col]==label_order[0]]['count'])
    x2, y2 = list(df_percentage[df_percentage[label_col]==label_order[1]][cat_col]), list(df_percentage[df_percentage[label_col]==label_order[1]]['count'])

    plt.subplot(121)
    ax1 = plt.bar(x1, y1, color= 'b', label=label_order[0])
    ax2 = plt.bar(x2, y2, color= 'r', bottom=y1, label=label_order[1])
    #plt.legend(loc=(1.05, 0.4))
    plt.ylabel('count')
    plt.grid(axis='x')

    plt.subplot(122)

    #stacked percentage barplot          
    x1, y1 = list(df_percentage[df_percentage[label_col]==label_order[0]][cat_col]), list(df_percentage[df_percentage[label_col]==label_order[0]]['%'])
    x2, y2 = list(df_percentage[df_percentage[label_col]==label_order[1]][cat_col]), list(df_percentage[df_percentage[label_col]==label_order[1]]['%'])

    ax1 = plt.bar(x1, y1, color= 'b', label=label_order[0])
    ax2 = plt.bar(x2, y2, color= 'r', bottom=y1, label=label_order[1])
    plt.legend(loc=(1.05, 0.4))
    plt.ylabel('%')
    plt.grid(axis='x')
```

```python
def dist_plot(data, category, continuous, kde_true, bins, colors, loc_legend):
    
    """
    Input: 
    1. data: pandas DataFrame.
    2. category: str, nombre de una columna categórica. Ej: 'sexo' = {'femenino', 'masculino'}
    3. continuous: str, nombre de columna continua. Ej: 'ingreso'
    4. kde_true: bool, si es True el histograma es normalizado (su área es 1), si es False histograma de frecuencias.
    5. bins: list, lista con los bins que tendrá el histograma de cada una de las categorías.
    6. loc_legend: tuple, localización de la legenda en el gráfico.
    
    Output:
    
    Retorna en una figura: boxplot horizontal e histogramas para difentes categorías.

    """
    
    #omitir nulos
    data = data[data[category].isnull()==False]
    #columnas de interés
    data = data[[category, continuous]]
    #resetear indices
    data.reset_index(drop=True, inplace=True)
    #categorías únicas
    categories = list(data[category].unique())
    
    #estilo
    sns.set(style="ticks")
   
    #se crean las figuras que contendran los gráficos
    f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)})
    
    #boxplot
    sns.boxplot(data=data, x=continuous, y=category, ax=ax_box, palette=colors)
    
    #histograma por cada categoría
    for i, cat in enumerate(categories):
        sns.distplot(data[(data[category]==cat)][continuous], bins=bins[i], color = colors[i], kde=kde_true, norm_hist=False, ax=ax_hist)
        
    #estilo
    ax_box.set(yticks=[])
    ax_box.set_xlabel('')
    ax_box.set_ylabel('')
    sns.despine(ax=ax_hist)
    sns.despine(ax=ax_box, left=True)
    plt.legend((categories), loc=1, bbox_to_anchor=loc_legend)
```

## 2.6 Gráfico Multidimensional

```python
df_parallel  = df.copy()
df_parallel['FUGA'] = df_parallel.apply(lambda x: 1 if x['FUGA']=='FUGA' else 0, axis=1)
df_parallel['GENERO'] = df_parallel.apply(lambda x: 1 if x['GENERO']=='M' else 0, axis=1)
```

```python
columns = ['GENERO', 'EDAD', 'M_MOROSO', 'MONTO','FUGA']
fig = px.parallel_coordinates(df_parallel[columns], color="FUGA", labels={"FUGA": "Fuga", "M_MOROSO": "Meses de mora",
        "MONTO": "Monto deuda","EDAD": "Edad", "GENERO": "Genero"},
        color_continuous_scale=px.colors.diverging.Tealrose, color_continuous_midpoint=0.5)
fig.show()
```

```python
df_parallel_mean_by_group = df_parallel.groupby('FUGA').mean()
df_parallel_mean_by_group.loc[:, 'FUGA'] = df_parallel_mean_by_group.index
df_parallel_mean_by_group
```

```python
columns = ['GENERO', 'EDAD', 'M_MOROSO', 'MONTO','FUGA']
fig = px.parallel_coordinates(df_parallel_mean_by_group[columns], color="FUGA", labels={"FUGA": "Fuga", "M_MOROSO": "Meses de mora",
        "MONTO": "Monto deuda","EDAD": "Edad",  "GENERO": "Genero"},
        color_continuous_scale=px.colors.diverging.Tealrose, color_continuous_midpoint=0.5)
fig.show()
```

Del parallel coordinates se observen los siguientes patrónes:
1. La fuga aumenta con los meses de mora y que solo hay personas de la clase no fuga en el grupo sin deudas (2.1)
2. La mayoría de los que registran altos montos de deuda son de la clase no fuga (2.2).
3. Las personas del grupo no fuga tienen una concentración mayor en el tramo más jovén que el grupo fuga el cual tiene una distribución mas o menos simétrica entorno a los 46 años (2.4).
